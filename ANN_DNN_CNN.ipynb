```python
import numpy as np
data = np.load(r'/content/drive/MyDrive/Colab Notebooks/dataset.npy')
label = np.load(r"/content/drive/MyDrive/Colab Notebooks/labels.npy")

test = np.load(r"/content/drive/MyDrive/Colab Notebooks/public_test.npy")
```

```python
label.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(data,label,test_size=0.35, random_state=42)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y_test)
test_labels_encoded = le.transform(y_test)
le.fit(y_train)
train_labels_encoded = le.transform(y_train)
```

```python
from keras.utils import to_categorical
y_train_one_hot = to_categorical(train_labels_encoded)
y_test_one_hot = to_categorical(test_labels_encoded)
```

```python
# tạo mô hình mạng nơ-ron tuần tự
model = keras.models.Sequential()
# theo anh nhớ thì feature vector là [1, 1024]
model.add(keras.layers.Input(shape=(2048,)))
# hidden layer 1 có 300 neuron
model.add(keras.layers.Dense(300, activation="relu"))
# hidden layer 2 có 100 neuron
model.add(keras.layers.Dense(100, activation="relu"))
# dùng activation softmax để phân loại 3 class
model.add(keras.layers.Dense(3, activation="softmax"))
```

```python
# các thông tin về trọng số của mô hình
model.summary()
```

```python
checkpoint = ModelCheckpoint("/content/drive/MyDrive/Colab Notebooks/ANN.h5", monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
# label của model theo anh nhớ là index của class nên dùng hàm loss "sparse_categorical_crossentropy"
# nếu như label biểu diễn ở dạng one-hot coding thì dùng hàm loss "categorical_crossentropy"
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
# train mô hình, ở đây anh train mô hình qua 30 epochs
# cái giá trị 0.1 chính là 10% dữ liệu được trích xuất ra từ training set để validation mô hình
# để xem thử có bị overfitting hay data mismatch trong lúc train hay không
history = model.fit(X_train, y_train_one_hot, validation_data=(X_test, y_test_one_hot), epochs=30, callbacks=[checkpoint])
```

```python
# đánh giá mô hình
model.evaluate(X_test, y_test_one_hot)

# # muốn predict dữ liệu mới thì dùng hàm predict
# model.predict(...)
```

## DNN

```python
import tensorflow as tf
from tensorflow import keras
from keras.callbacks import ModelCheckpoint
```

```python
batch_size = 128
epochs = 100
learning_rate = 0.006
```

```python
# tạo mạng nơ ron tuần tự
model = keras.Sequential()

# input layer
model.add(keras.layers.InputLayer(input_shape=[2048]))


# hidden layer 1
model.add(keras.layers.Dense(1024))
model.add(keras.layers.LeakyReLU(alpha=0.01))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Dropout(0.2))

# hidden layer 3
model.add(keras.layers.Dense(256))
model.add(keras.layers.LeakyReLU(alpha=0.01))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Dropout(0.2))

# hidden layer 5
model.add(keras.layers.Dense(64))
model.add(keras.layers.LeakyReLU(alpha=0.01))


# output layer
model.add(keras.layers.Dense(3, activation='softmax'))

# compile mô hình
# label của model theo anh nhớ là index của class nên dùng hàm loss "sparse_categorical_crossentropy"
# nếu như label biểu diễn ở dạng one-hot coding thì dùng hàm loss "categorical_crossentropy"
checkpoint = ModelCheckpoint("/content/drive/MyDrive/Colab Notebooks/DNN.h5", monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')

model.compile(loss="categorical_crossentropy", optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate) , metrics=["accuracy"])
# train mô hình, ở đây anh train mô hình qua 20 epochs
# cái giá trị 0.1 chính là 10% dữ liệu được trích xuất ra từ training set để validation mô hình
# để xem thử có bị overfitting hay data mismatch trong lúc train hay không
history = model.fit(X_train, y_train_one_hot, validation_data=(X_test, y_test_one_hot), epochs=epochs, batch_size=batch_size, callbacks=[checkpoint])
```

```python
# đánh giá mô hình
model.evaluate(X_test, y_test_one_hot)
```

```python
import re
from sentence_transformers import SentenceTransformer, util
from tqdm import tqdm
import json
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

result_dict = {}

model = SentenceTransformer("sentence-transformers/paraphrase-xlm-r-multilingual-v1")

ohe_vectorizer = CountVectorizer(binary=True)

tfidf_vectorizer = TfidfVectorizer()

# Loop through each item in the dataset with tqdm for progress tracking
for key, data in tqdm(dataset.items(), desc="Processing Items"):
    # Extract context and claim from the current item
    context = data['context']
    claim = data['claim']
    verdict = data['verdict']

    # Split context into sentences
    sentences = re.split(r'\. |\n\n|(?<=\.”)', context)

    # Compute embeddings for claim using Sentence Transformer
    claim_embedding = model.encode(claim, convert_to_tensor=True)

    # One-Hot Encoding (OHE) for claim and context sentences
    claim_ohe = ohe_vectorizer.transform([claim]).toarray()
    context_ohe = ohe_vectorizer.transform(sentences).toarray()

    # TF-IDF for claim and context sentences
    claim_tfidf = tfidf_vectorizer.transform([claim]).toarray()
    context_tfidf = tfidf_vectorizer.transform(sentences).toarray()

    if verdict != "NEI":
        max_combined_score = -1
        max_score_sentence = ""
        for i, sentence in enumerate(sentences):
            # Calculate cosine similarity between embeddings
            cosine_sim = util.pytorch_cos_sim(claim_embedding, model.encode(sentence, convert_to_tensor=True)).item()

            ohe_cosine_sim = cosine_similarity(claim_ohe, context_ohe[i].reshape(1, -1))[0][0]

            tfidf_cosine_sim = cosine_similarity(claim_tfidf, context_tfidf[i].reshape(1, -1))[0][0]

            # Combine the similarities (you can adjust the weights)
            combined_sim = 0.5 * cosine_sim + 0.25 * ohe_cosine_sim + 0.25 * tfidf_cosine_sim

            if combined_sim > max_combined_score:
                max_combined_score = combined_sim
                max_score_sentence = sentence
    else:
        max_score_sentence = ""

    # Store the results in the result_dict
    result_dict[key] = {
        "verdict": verdict,
        "evidence": max_score_sentence  # Set evidence to the highest-scoring sentence
    }

# Save the results to a JSON file
with open('public_result11.json', 'w', encoding='utf-8') as output_file:
    json.dump(result_dict, output_file, ensure_ascii=False, indent=2)
```
